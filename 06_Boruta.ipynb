__Import packages__

```python
from boruta import BorutaPy
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

__Make Dataset__

```python
import pandas as pd
from mrmr import mrmr_classif
from sklearn.datasets import make_classification

# create some data
X, y = make_classification(n_samples = 1000, 
                           n_features = 50, 
                           n_informative = 10, 
                           n_redundant = 40)
```

__Feature Selection with Boruta__

```python
# avoid deprecation error
np.int = np.int32
np.float = np.float64
np.bool = np.bool_
```

```python

# Initialize Boruta
forest = RandomForestRegressor(
   n_jobs = -1, 
   max_depth = 5
)


# Boruta using RF
boruta = BorutaPy(
   estimator = forest, 
   n_estimators = 'auto',
   max_iter = 100 # number of trials to perform
)

# Only np.array is accepted
boruta.fit(np.array(X), np.array(y))
```

__Accepted and tenatative features__

```python

# print results
accepted = X[:, boruta.support_]
tentative = X[:, boruta.support_weak_]
print('features in the accepted area:', accepted.shape)
print('features in the tentative area:', tentative.shape)
```
